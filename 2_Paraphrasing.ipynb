```python
from parrot import Parrot
import torch
import warnings
warnings.filterwarnings("ignore")
```

```python
def random_state(seed):
  torch.manual_seed(seed)
  if torch.cuda.is_available():
    torch.cuda.manual_seed_all(seed)

random_state(1234)
```

```python
parrot = Parrot(model_tag="prithivida/parrot_paraphraser_on_T5", use_gpu=False)
```

```python
phrases = ["What's the most delicious papayas?"]
```

```python
def paraphrase(input):
   result = translator.translate(input, src='hi', dest='en')
   para_phrases = parrot.augment(input_phrase=result.text)
   for para_phrase in para_phrases:
     x=translator.translate(para_phrase,src='en',dest='hi')
     print(x.text)
```

```python
result = translator.translate('क्या आप कॉल बैक सुविधा प्रदान करते हैं', src='hi', dest='en')
```

```python
paraphrase("क्या आप कॉल बैक सुविधा प्रदान करते हैं")
```
